Text to image converter

```python
from diffusers import StableDiffusionPipeline
import torch

# Load the pre-trained Stable Diffusion model from Hugging Face
pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5")
pipe = pipe.to("cuda")  # Move model to GPU if available

# Input text to convert to an image
text = "A beautiful sunset over the mountains with a river flowing through the valley."

# Generate the image from the text
image = pipe(text).images[0]

# Save and show the image
image.save("generated_image.png")
image.show()
```
