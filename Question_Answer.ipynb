```python
from transformers import pipeline

# Initialize a smaller, faster question-answering pipeline
pipe = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")

# Step 1: Define the context paragraph
paragraph = """
The Amazon rainforest is one of the most biodiverse regions on Earth, home to millions of species of plants, animals, and insects. It plays a critical role in regulating the global climate by absorbing vast amounts of carbon dioxide. However, deforestation and other human activities threaten this vital ecosystem, impacting not only the species within but also contributing to global warming.
"""

# Step 2: Define the question
question = "Why is the Amazon rainforest important for the global climate?"

# Step 3: Use the pipeline to generate an answer
response = pipe(question=question, context=paragraph)

# Print the generated answer
print("Answer:", response["answer"])
```
